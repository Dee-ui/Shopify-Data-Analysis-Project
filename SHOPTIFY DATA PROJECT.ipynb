import numpy as np
import random
# 1. Introduce Missing Values
# Randomly select 5% of the dataset and replace with NaN for specific columns
cols_with_missing_values = ['explicit', 'danceability', 'energy', 'acousticness', 'liveness']
for col in cols_with_missing_values:
    df.loc[df.sample(frac=0.05).index, col] = np.nan

# 2. Add Duplicate Rows
# Randomly duplicate 3% of the rows
duplicate_count = int(0.03 * len(df))
duplicates = df.sample(n=duplicate_count)
df = pd.concat([df, duplicates])

# 3. Inconsistent Data Formats

# Modify 'currency' to have inconsistent casing
df['artist'] = df['artist'].apply(lambda x: x.lower() if random.random() > 0.5 else x.upper())

# Modify 'product_price' to insert currency symbols inconsistently
df['year'] = df['year'].apply(lambda x: f'${x}' if random.random() > 0.5 else x)

# 4. Add Outliers or Erroneous Data
# Add unrealistic product prices (negative or very high values)
outlier_count = int(0.01 * len(df))
outliers = df.sample(n=outlier_count).index
df.loc[outliers, 'key'] = df.loc[outliers, 'key'].apply(lambda x: random.choice([-999.99, 99999.99]))

# Add negative quantities or extremely high values
df.loc[outliers, 'mode'] = df.loc[outliers, 'mode'].apply(lambda x: random.choice([-5, 10000]) if pd.notna(x) else x)

# 5. Add Whitespace and Extra Characters
# Add random whitespace to string columns like 'product_name'
df['artist'] = df['artist'].apply(lambda x: f' {x} ' if pd.notna(x) and random.random() > 0.5 else x)

import random

# List of special characters to randomly add
special_chars = ['#', '@', '$', '%', '&', '*', '!', '^']

# Function to randomly add three special characters
def add_special_chars(x):
    chars_to_add = ''.join(random.choices(special_chars, k=3))  # Select 3 random characters
    return f'{x}{chars_to_add}' if random.random() > 0.5 else f'{chars_to_add}{x}'

# Add three special characters randomly to 'customer_id' and 'device_id'
df['tempo'] = df['tempo'].apply(add_special_chars)
df['genre'] = df['genre'].apply(add_special_chars)
